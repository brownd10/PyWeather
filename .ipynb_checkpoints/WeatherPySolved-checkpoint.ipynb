```python
# Create dependencies to be used for file script
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import openweathermapy.core as owm
import requests
import json
import random
import time
import seaborn as sns
from citipy import citipy

# OpenWeather API key
from config import wkey
```

```python
# Read in csv file
cities_pd = pd.read_csv("worldcities.csv")

# Preview the data
cities_pd.head()
```

```python
# Randomly select at least 500 cities that have a latitude within 
selected_cities = cities_pd.sample(n=550)
selected_cities = selected_cities[selected_cities["Latitude"].astype(float) != 0]

# Visualize dataset
selected_cities.count()
```

```python
selected_cities.infer_objects().dtypes
```

```python
# View the sampled dataframe
selected_cities.head()
```

```python
# Create blank columns for necessary fields
selected_cities["Temperature (F)"] = ""
selected_cities["Humidity (%)"] = ""
selected_cities["Cloudiness (%)"] = ""
selected_cities["Wind Speed (mph)"] = ""

# Initiate counter
row_count = 0

# loop through an grab the lat/lng using Google maps
for index, row in selected_cities.iterrows():
    
    main_url = "http://api.openweathermap.org/data/2.5/weather?q=" + row['City'] + ',' + row['Country'] + "&units=imperial" + "&appid=" + wkey
    
    # Loop log for error handling
    print("Now retrieving city # " + str(row_count))
    row_count += 1
    
    # Run requests to grab the JSON at the requested URL
    city_location = requests.get(main_url)
    print(city_location.url)
    
    city_location = city_location.json()
    
    # Append the blank columns to the dataset
    try:
        city_temp = city_location["main"]["temp"]
        city_humid = city_location["main"]["humidity"]
        city_cloud = city_location["clouds"]["all"]
        city_wind = city_location["wind"]["speed"]
        
        selected_cities.set_value(index, "Temperature (F)", city_temp)
        selected_cities.set_value(index, "Humidity (%)", city_humid)
        selected_cities.set_value(index, "Cloudiness (%)", city_cloud)
        selected_cities.set_value(index, "Wind Speed (mph)", city_wind)
        
    except (KeyError, IndexError):
        print("There is an ERROR with this city data. Skipping!")
        continue
        
# Preview the data
selected_cities.head()
```

```python
selected_cities.convert_objects(convert_numeric=True).dtypes
```

```python
# Create temperature scatter plot
plt.scatter(selected_cities['Latitude'], selected_cities['Temperature (F)'], marker="o", facecolors="cyan", edgecolors="black", alpha=0.50)

plt.title("City Latitude vs Temperature " + "(" + time.strftime("%m/%d/%Y") + ")", fontsize = 18)
plt.xlabel("Latitude", fontsize = 14)
plt.xlim(-90, 90)
plt.ylabel("Temperature (F)", fontsize = 14)
plt.ylim(-50, 400)

# Save the plot as a PNG file
plt.savefig("Latitude_vs_Temperature.png")

plt.grid()
plt.yticks([])
plt.show()
```

```python
# Create humidity scatter plot
plt.scatter(selected_cities['Latitude'], selected_cities['Humidity (%)'], marker="o", facecolors="blue", edgecolors="grey", alpha=0.50)

plt.title("City Latitude vs Humidity " + "(" + time.strftime("%m/%d/%Y") + ")", fontsize = 18)
plt.xlabel("Latitude", fontsize = 14)
plt.xlim(-90, 90)
plt.ylabel("Humidity (%)", fontsize = 14)
plt.ylim(-50, 150)

# Save the plot as a PNG file
plt.savefig("Latitude_vs_Humidity.png")

plt.grid()
plt.yticks([])
plt.show()
```

```python
# Create cloudiness scatter plot
plt.scatter(selected_cities['Latitude'], selected_cities['Cloudiness (%)'], marker="o", facecolors="grey", edgecolors="magenta", alpha=0.50)

plt.title("City Latitude vs Cloudiness " + "(" + time.strftime("%m/%d/%Y") + ")", fontsize = 18)
plt.xlabel("Latitude", fontsize = 14)
plt.xlim(-90, 90)
plt.ylabel("Cloudiness (%)", fontsize = 14)
plt.ylim(-10, 45)

# Save the plot as a PNG file
plt.savefig("Latitude_vs_Cloudiness.png")

plt.grid()
plt.yticks([])
plt.show()
```

```python
# Create cloudiness scatter plot
plt.scatter(selected_cities['Latitude'], selected_cities['Wind Speed (mph)'], marker="o", facecolors="red", edgecolors="black", alpha=0.50)

plt.title("City Latitude vs Wind Speed " + "(" + time.strftime("%m/%d/%Y") + ")", fontsize = 18)
plt.xlabel("Latitude", fontsize = 14)
plt.xlim(-90, 90)
plt.ylabel("Wind Speed (mph)", fontsize = 14)
plt.ylim(-10, 150)

# Save the plot as a PNG file
plt.savefig("Latitude_vs_Wind_Speed.png")

plt.grid()
plt.yticks([])
plt.show()
```
